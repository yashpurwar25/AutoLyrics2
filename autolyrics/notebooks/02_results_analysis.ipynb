```python
# Cell 1 — Load results
import json, pandas as pd, matplotlib.pyplot as plt, seaborn as sns

results = json.load(open("results/eval_report.json"))
df = pd.DataFrame(results).T.reset_index().rename(columns={"index": "model"})

# Cell 2 — WER bar chart
fig, ax = plt.subplots(figsize=(8, 4))
colors = ["#fc5c5c", "#fcdc3c", "#3cfca0"]
bars = ax.bar(df["model"], df["wer"], color=colors, width=0.5)
ax.axhline(y=df.iloc[0]["wer"] * 0.85, color="white", ls="--", lw=1.5,
           label="−15 % relative target")
ax.set_ylabel("WER (%)"); ax.set_title("AutoLyrics — WER Comparison")
ax.legend(); plt.tight_layout(); plt.savefig("results/wer_comparison.png", dpi=150)

# Cell 3 — Relative improvement table
df[["model", "wer", "cer", "wer_relative_improvement_%"]].style\
    .background_gradient(subset=["wer"], cmap="RdYlGn_r")\
    .format({"wer": "{:.2f}", "cer": "{:.2f}", "wer_relative_improvement_%": "{:+.1f}"})
```
